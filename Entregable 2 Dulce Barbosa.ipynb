# $\textrm{Tabla Sprint Results}$

Después de importar las paqueterias creamos cada tabla


```python
#Primero analizamos Sprint y que elementos contiene 

url='http://ergast.com/api/f1/sprint.json'
response = requests.get(url.format(2022))
content=json.loads(response.content)
content['MRData']['RaceTable']['Races'][0]['SprintResults'][0].keys()
```

```python
#Creamos el diccionario
sprint = {'RaceId':[],
        'DriverId':[],
        'ConstructorId':[],
        'Number':[],
        'Position':[],
        'PositionText':[],
        'Points':[],
        'Laps': [],
        'Grid':[],
        'Time':[], 
        'Millis':[],
        'Fastestlap':[],
        'Fastlaptime':[],
        'Status':[]
       }
```

```python
#Creamos la función
for year in list(range(2010,2023)):
    url ='http://ergast.com/api/f1/sprint.json'
    response = requests.get(url.format(year))
    
    content = json.loads(response.content)
        
    for i in content['MRData']['RaceTable']['Races'][0]['SprintResults']:
        sprint['RaceId'].append(content['MRData']['RaceTable']['Races'][0]['raceName'])
        sprint['DriverId'].append(i['Driver']['driverId'])
        sprint['ConstructorId'].append(i['Constructor']['constructorId'])
        sprint['Number'].append(int(i['number']))
        sprint['Position'].append(int(i['position']))
        sprint['PositionText'].append(i['positionText'])
        sprint['Laps'].append(i['laps'])
        sprint['Grid'].append(i['grid'])
        sprint['Points'].append(int(i['points']))
        sprint['Status'].append(i['status'])
        
        try:
            sprint['Millis'].append(i['Time']['millis'])
        except:
             sprint['Millis'].append(None)
        try:
            sprint['Time'].append(i['Time']['time'])
        except:
            sprint['Time'].append(None)
        try:
            sprint['Fastestlap'].append(i['FastestLap']['lap'])
        except:
            sprint['Fastestlap'].append(None)
        try:
            sprint['Fastlaptime'].append(i['FastestLap']['Time']['time'])
        except:
            sprint['Fastlaptime'].append(None)
            
    for i in content['MRData']['RaceTable']['Races'][1]['SprintResults']:
        sprint['RaceId'].append(content['MRData']['RaceTable']['Races'][1]['raceName'])
        sprint['DriverId'].append(i['Driver']['driverId'])
        sprint['ConstructorId'].append(i['Constructor']['constructorId'])
        sprint['Number'].append(int(i['number']))
        sprint['Position'].append(int(i['position']))
        sprint['PositionText'].append(i['positionText'])
        sprint['Laps'].append(i['laps'])
        sprint['Grid'].append(i['grid'])
        sprint['Points'].append(int(i['points']))
        sprint['Status'].append(i['status'])
        
        try:
            sprint['Millis'].append(i['Time']['millis'])
        except:
             sprint['Millis'].append(None)
        try:
            sprint['Time'].append(i['Time']['time'])
        except:
            sprint['Time'].append(None)
        try:
            sprint['Fastestlap'].append(i['FastestLap']['lap'])
        except:
            sprint['Fastestlap'].append(None)
        try:
            sprint['Fastlaptime'].append(i['FastestLap']['Time']['time'])
        except:
            sprint['Fastlaptime'].append(None)
```

```python
pd.DataFrame(sprint).head()
```

```python
#Creamos el archivo csv
pd.DataFrame(sprint).to_csv('Sprint.csv', index=False)
```

# $\textrm{Tabla Status}$

```python
#Primero analizamos Status y que elementos contiene 

url='http://ergast.com/api/f1/status.json'
response = requests.get(url.format(2022))
content=json.loads(response.content)
content['MRData']['StatusTable']['Status'][0].keys()
```

```python
#Creamos el diccionario 

status = {'StatusId':[],
        'Count':[],
        'Status':[]
       }
```

```python
#Creamos la función

for year in list(range(2010,2023)):
    url ='http://ergast.com/api/f1/status.json'
    response = requests.get(url.format(year))
    
    content = json.loads(response.content)
        
    for i in content['MRData']['StatusTable']['Status']:
        status['StatusId'].append(int(i['statusId']))
        status['Count'].append(int(i['count']))
        status['Status'].append(i['status'])
```

```python
pd.DataFrame(status).head()
```

```python
#Importamos a csv

pd.DataFrame(status).to_csv('Status.csv', index=False)
```

# $\textrm{Tabla Pit Stops}$

```python
url='http://ergast.com/api/f1/2022/5/pitstops.json'
response = requests.get(url.format())
content=json.loads(response.content)
content['MRData']['RaceTable']['Races'][0]['PitStops'][0].keys()
```

```python
#Creamos el diccionario

PitStops = {'RaceId':[],
        'DriverId':[],
        'Lap':[],
        'Stop':[],
        'Time':[],
       }
```

```python
#Creamos la función 

for year in range(2011,2023):
    url ='http://ergast.com/api/f1/{}/5/pitstops.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    
    for  i in content['MRData']['RaceTable']['Races'][0]['PitStops']:
        PitStops['RaceId'].append(content['MRData']['RaceTable']['Races'][0]['raceName'])
        PitStops['DriverId'].append(i['driverId'])
        PitStops['Lap'].append(i['lap'])
        PitStops['Stop'].append(i['stop'])
        PitStops['Time'].append(i['time'])
        PitStops['Duration'].append(i['duration'])

        
```
